***Ranking and Weight Calculation***

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
csv_file = "/content/rawdomCasmurroUpdate.csv"
df = pd.read_csv(csv_file, encoding='latin1')

data = df["English"].tolist()
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(data)
print("vectorizer: ",vectorizer)
print("tfidf_matrix: ",tfidf_matrix)
```

```python
row_sums = np.array(tfidf_matrix.sum(axis=1)).flatten()
row_sums[row_sums == 0] = 1
normalized_tfidf = tfidf_matrix / row_sums[:, np.newaxis]

# Calculate term weights using AHP
term_weights = normalized_tfidf.sum(axis=0)
print("Term Weights:")
print(term_weights)
```

```python
ranks = term_weights.argsort()[::-1]
print("Ranks: ",ranks)
```

***Top-ranked keywords***

```python
num_terms = 10
vocabulary = vectorizer.vocabulary_
top_keywords = [term for term, idx in vocabulary.items() if idx in ranks[:num_terms]]

# Print the top-ranked keywords
print("Top-ranked Keywords:")
for keyword in top_keywords:
    print(keyword)
```

```python
from nltk.corpus import wordnet
from nltk.wsd import lesk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

***WSD (Word Sense Disambiguation)***

```python
def perform_wsd(sentence):
    lemmatizer = WordNetLemmatizer()
    tokens = sentence.lower().split()
    wsd_results = []
    for token in tokens:
        synset = lesk(tokens, token)
        if synset is not None:
            lemma = lemmatizer.lemmatize(token)
            wsd_results.append((token, lemma, synset.definition()))
    return wsd_results

sentence = "the eyes three or four times"
wsd_results = perform_wsd(sentence)
print("WSD Results:")
for result in wsd_results:
    print("Token: ", result[0])
    print("Lemma: ", result[1])
    print("Definition: ", result[2])
    print()
```

***Query Expansion***



```python
def expand_query(query_terms):
    expanded_terms = []
    lemmatizer = WordNetLemmatizer()
    for term in query_terms:
        tokens = word_tokenize(term)
        lemmas = [lemmatizer.lemmatize(token) for token in tokens]
        for lemma in lemmas:
            synsets = wordnet.synsets(lemma)
            if synsets:
                related_terms = [synset.lemmas()[0].name() for synset in synsets]
                expanded_terms.extend(related_terms)
    return expanded_terms
query_terms = ["He greeted me, sat down next to me", "spoke of the Moon and the ministers"]
expanded_terms = expand_query(query_terms)
print("Expanded Terms:")
print(expanded_terms)
```

```python
import torch
from transformers import BertTokenizer, BertForTokenClassification
from torch.utils.data import TensorDataset, DataLoader
from torch import nn, optim
from tqdm import tqdm
```

```python
tokens = []
label_ids = []
label_map = {
    "label_1": 1,
    "label_2": 2,
    "label_3": 3,
}
for sentence, label in zip(sentences, labels):
    encoded_sentence = tokenizer.encode(sentence, add_special_tokens=True)
    encoded_labels = [label_map.get(label, 0)] + [0] * (len(encoded_sentence) - 1)
    tokens.extend(encoded_sentence)
    label_ids.extend(encoded_labels)
```

```python
max_length = 32
padding_length = max_length - len(tokens)
if padding_length > 0:
    tokens += [tokenizer.pad_token_id] * padding_length
    label_ids += [0] * padding_length
elif padding_length < 0:
    tokens = tokens[:max_length]
    label_ids = label_ids[:max_length]
```

```python
tokens_tensor = torch.tensor(tokens).unsqueeze(0)
label_ids_tensor = torch.tensor(label_ids).unsqueeze(0)
dataset = TensorDataset(tokens_tensor, label_ids_tensor)
dataloader = DataLoader(dataset, batch_size=1)
loss_function = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader
from transformers import BertForTokenClassification, AdamW
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
label_ids_tensor = torch.tensor(label_ids)
tokens_tensor = torch.tensor(tokens)
attention_mask_tensor = torch.tensor([1 if token != 0 else 0 for token in tokens])

batch_size = 32
learning_rate = 1e-5

dataset = TensorDataset(tokens_tensor, attention_mask_tensor, label_ids_tensor)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

model = BertForTokenClassification.from_pretrained('bert-base-uncased', num_labels=num_labels)
optimizer = AdamW(model.parameters(), lr=learning_rate)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
num_epochs = 2
model.to(device)  # Move the model to the same device
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    total_accuracy = 0
    total_precision = 0
    total_recall = 0
    total_f1 = 0
    total_steps = 0

    for batch in dataloader:
        optimizer.zero_grad()
        batch = tuple(t.to(device) for t in batch)
        tokens_tensor = batch[0].unsqueeze(0).to(device)
        attention_mask_tensor = batch[1].unsqueeze(0).to(device)
        labels_tensor = batch[2].unsqueeze(0).to(device)
        outputs = model(input_ids=tokens_tensor, attention_mask=attention_mask_tensor, labels=labels_tensor)
        loss = outputs.loss
        logits = outputs.logits

        _, predicted_labels = torch.max(logits, dim=2)
        flat_predicted_labels = predicted_labels.view(-1).cpu().numpy()
        flat_label_ids = labels_tensor.view(-1).cpu().numpy()
        accuracy = accuracy_score(flat_label_ids, flat_predicted_labels)
        precision = precision_score(flat_label_ids, flat_predicted_labels, average='weighted')
        recall = recall_score(flat_label_ids, flat_predicted_labels, average='weighted')
        f1 = f1_score(flat_label_ids, flat_predicted_labels, average='weighted')

        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        total_accuracy += accuracy
        total_precision += precision
        total_recall += recall
        total_f1 += f1
        total_steps += 1

    average_loss = total_loss / total_steps
    average_accuracy = total_accuracy / total_steps
    average_precision = total_precision / total_steps
    average_recall = total_recall / total_steps
    average_f1 = total_f1 / total_steps

    print(f"Epoch {epoch+1}: Loss = {average_loss}, Accuracy = {average_accuracy}, Precision = {average_precision}, Recall = {average_recall}, F1 Score = {average_f1}")
```

***Retrieve the relevant documents***

```python
import pandas as pd
import torch
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertTokenizer, BertModel

data = pd.read_csv('/content/rawdomCasmurroUpdate.csv', encoding='latin-1')
queries = data['Portuguese'].tolist()
documents = data['English'].tolist()
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
query_tokens = tokenizer.batch_encode_plus(
    queries,
    truncation=True,
    padding=True,
    max_length=128,
    return_tensors='pt'
)
document_tokens = tokenizer.batch_encode_plus(
    documents,
    truncation=True,
    padding=True,
    max_length=128,
    return_tensors='pt'
)
```

```python
query_input_ids = query_tokens['input_ids']
query_attention_mask = query_tokens['attention_mask']
document_input_ids = document_tokens['input_ids']
document_attention_mask = document_tokens['attention_mask']
with torch.no_grad():
    query_embeddings = model(
        input_ids=query_input_ids,
        attention_mask=query_attention_mask
    ).last_hidden_state[:, 0, :]
    document_embeddings = model(
        input_ids=document_input_ids,
        attention_mask=document_attention_mask
    ).last_hidden_state[:, 0, :]
```

```python
top_k = 5

relevant_documents = []
for i in range(len(queries)):
    query = queries[i]
    query_embedding = query_embeddings[i]
    similarity_scores = cosine_similarity(query_embeddings, document_embeddings)
    similarity_scores_tensor = torch.tensor(similarity_scores)
    sorted_indices = torch.argsort(similarity_scores_tensor, descending=True)
    relevant_docs = []
    for j in range(top_k):
        doc_idx = sorted_indices[0][j].item()
        doc_text = documents[doc_idx]
        relevance_score = similarity_scores[0][doc_idx].item()
        relevant_docs.append((doc_text, relevance_score))
    relevant_documents.append((query, relevant_docs))
for query, docs in relevant_documents:
    print(f"Query: {query}")
    for doc, score in docs:
        print(f"Document: {doc}")
        print(f"Relevance Score: {score}")
    print()
```
